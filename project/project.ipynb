```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,LabelBinarizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
scalar = StandardScaler()
encoder = LabelBinarizer(neg_label=-1, pos_label=1, sparse_output=False)
response = data['diagnosis']
y = encoder.fit_transform(response)
feature = data.drop(['id', 'diagnosis'], axis=1)
X = scalar.fit_transform(feature)
norm_data = pd.concat([pd.DataFrame(X, columns = feature.columns.values), 
                       pd.DataFrame(y, columns = ['diagnosis'])],
                       axis = 1)

norm_data.head(25)
```

```python
norm_data.corr()['diagnosis']
norm_data = norm_data.drop(['fractal_dimension_mean','texture_se','smoothness_se','symmetry_se'],axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
lr = LogisticRegression()
lr = lr.fit(X_train,y_train.ravel())
lr.score(X_test,y_test)
```
